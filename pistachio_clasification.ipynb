```python
df.describe()
```

```python
#count of unique individuals present in the column
df.Class.value_counts()
```

```python
#LABEL ENCODING

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Class']=le.fit_transform(df['Class'])
```

```python
df['Class']
```

```python
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
y.value_counts()
```

```python
sns.kdeplot(x['Area'])
```

```python
plt.title('ROUNDNESS OF TWO CLASSES')
sns.barplot(x='Class',y='Roundness',data=df)
plt.show()
```

```python
#Data balancing

from sklearn.ensemble import IsolationForest
iso=IsolationForest(random_state=0,contamination=0.05)
clean=iso.fit_predict(x,y)
```

```python
x=x[clean==1]
y=y[clean==1]
```

```python
x.shape
```

```python
y.value_counts()
```

```python
sns.countplot(x=y)
```

```python
#Dimensionality Reduction
x_new=x[['Area','Perimeter','Major_Axis','Minor_Axis','Roundness']]
```

```python
x_new
```

```python
x_new.shape
```

```python

#feature scaling

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x_new)
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
#MODEL BUILDING-LOGISTIC REGRESSION
#1.splitting data into train test
#2.logistic regression model
#3.confusion matrix analysis
#4.classification report

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    x_new,y,test_size=0.20,random_state=16
)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

LOGISTIC REGRESION

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression(random_state=16)
```

```python
logreg.fit(x_train,y_train)
```

```python
y_pred=logreg.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
cof_matrix=metrics.confusion_matrix(y_test,y_pred)
cof_matrix
```

```python
# import required modules

class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cof_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import classification_report
target_names = ['Kirmizi_Pistachio','Siirt_Pistachio']
print(classification_report(y_test, y_pred, target_names=target_names))
```

DECISION CLASSIFIER

```python
#CHECKING THE ACCURACY OF THE MODEL USING DECISION CLASSIFIER ALGORITHM

from sklearn.tree import DecisionTreeClassifier
x_train, x_test, y_train, y_test = train_test_split(x_new, y, random_state=0)
```

```python
#building a model

classifier = DecisionTreeClassifier(random_state=0)
classifier.fit(x_train, y_train)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(classifier);
```

```python
plt.figure(figsize=(12,12))
plot_tree(classifier, fontsize=7, filled=True, rounded=True,
         feature_names=list(x.columns), class_names=['0','1']);
```

```python
#performance Evaluation
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score, classification_report
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
pred=pd.DataFrame(
{
    'actual data': y_test,
    'predicted data': y_pred
})
pred
```
